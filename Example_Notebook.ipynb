```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Copa_america.csv") 

Country=data.Nationality.unique()
```

```python
for i, pais in enumerate(Country):
    Pais=data[data["Nationality"] == str(pais)]
    Edades=Pais.Age
    plt.title("Edad Promedio de "+str(pais)+ " es "+str(int(Pais.Age.mean()))+ " años y " + str(int(Pais.Age.mean()%1*12)) +" meses" )
    plt.hist(Edades, bins= 10)
    plt.axvline(Pais.Age.mean(), color='k', linestyle='dashed', linewidth=1)
    plt.show()
```
